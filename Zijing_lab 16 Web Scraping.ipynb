####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
date_time = [d.text.strip().split('\xa0\xa0\xa0') for d in soup.find_all('td',{'class':"tabev6"})]
for i in (0,40):
    date=date_time[i]
    
    
region = [r.text.strip() for r in soup.find_all('td',{'class':"tb_region"})]

#print(latitude_longitude)

latitude = []
longitude = []

ll_list = [l.text.strip() for l in soup.find_all('td',{'class':"tabev1"})]
#print(len(ll_list))

for i in range(0,100):
    a = ll_list[i]
    if i%2 == 0:
        latitude.append(a)
    else:
        longitude.append(a)

print(latitude)
````

#### Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here
html = requests.get(url).content
soup = BeautifulSoup(html,"html.parser")
soup
````

````python
movie = soup.find_all('td', {'class':"titleColumn"})
#movie
m_name = [m.text.replace('\n',' ').strip() for m in soup.find_all('td',{'class':"titleColumn"})]
m_name
````

````python
#d_name = [d.find('a')['title'] for d in movie]
d_name = [m.find('a')['title'].replace("(dir.)", "").split(',')[0] for m in movie]
#d_name
a_name = [m.find('a')['title'].split(',')[1:] for m in movie]
#a_name
````

````python
#top 250 data (movie name, initial release, director name and stars) 
cols = ['Title(Year)', 'Director', 'Stars']

data = []
for i in range(250):
    data.append([m_name[i], d_name[i], (",".join(a_name[i]))])

result =pd.DataFrame(data,columns=cols)
result
````
